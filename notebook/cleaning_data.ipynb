````python
import pandas as pd
import numpy as np
from IPython.display import display
````

````python

datasets = {
    "hotel_bookings": "../data/raw/hotel_bookings.csv",
    "customers": "../data/raw/olist_customers_dataset.csv",
    "geolocation": "../data/raw/olist_geolocation_dataset.csv",
    "order_items": "../data/raw/olist_order_items_dataset.csv",
    "payments": "../data/raw/olist_order_payments_dataset.csv",
    "reviews": "../data/raw/olist_order_reviews_dataset.csv",
    "orders": "../data/raw/olist_orders_dataset.csv",
    "products": "../data/raw/olist_products_dataset.csv",
    "sellers": "../data/raw/olist_sellers_dataset.csv",
    "category_translation": "../data/raw/product_category_name_translation.csv"
}



dfs = {}

for name, path in datasets.items():
    dfs[name] = pd.read_csv(path)

def analyze_dataset(name, df):

    print("\n" + "="*120)
    print(f"DATASET: {name.upper()}")
    print("="*120)


    print("\n[1] SAMPLE DATA (10 ROWS)")

    sample_size = min(10, len(df))
    display(df.sample(sample_size))


    print("\n[2] DATA SHAPE")

    shape_df = pd.DataFrame({
        "Rows": [df.shape[0]],
        "Columns": [df.shape[1]]
    })

    display(shape_df)


    print("\n[3] DATA TYPES")

    dtype_df = pd.DataFrame({
        "Column": df.columns,
        "Data Type": df.dtypes.astype(str)
    })

    display(dtype_df)


    print("\n[4] DESCRIPTIVE STATISTICS")

    display(df.describe(include="all").T)

    print("\n[5] MISSING VALUES")

    missing_df = pd.DataFrame({
        "Missing Count": df.isnull().sum(),
        "Missing Percentage (%)":
            round(df.isnull().mean() * 100, 2)
    })

    missing_df = (
        missing_df
        .sort_values(
            by="Missing Percentage (%)",
            ascending=False
        )
    )

    display(missing_df)

    print("\n[6] DUPLICATED ROWS")

    duplicate_count = df.duplicated().sum()

    duplicate_df = pd.DataFrame({
        "Duplicate Rows": [duplicate_count],
        "Duplicate Percentage (%)": [
            round(
                duplicate_count / len(df) * 100,
                2
            )
        ]
    })

    display(duplicate_df)


    print("\n[7] CARDINALITY")

    cardinality_df = pd.DataFrame({
        "Unique Values": df.nunique()
    })

    cardinality_df = (
        cardinality_df
        .sort_values(
            by="Unique Values",
            ascending=False
        )
    )

    display(cardinality_df)


for name, df in dfs.items():
    analyze_dataset(name, df)
````

## Why Use .copy() When Creating DataFrames?

Ban đầu dữ liệu được gán như sau:

```python
hotel_df = dfs["hotel_bookings"]
```

Cách gán này không tạo ra một DataFrame mới mà chỉ tạo một tham chiếu (reference) tới cùng một đối tượng DataFrame trong bộ nhớ.

Do đó, mọi thay đổi thực hiện trên `hotel_df` sẽ đồng thời làm thay đổi `dfs["hotel_bookings"]`.

Ví dụ:

```python
hotel_df.drop(columns=["company"], inplace=True)
```

Sau lệnh trên, cột `company` cũng bị xóa khỏi `dfs["hotel_bookings"]`.

Để tránh làm thay đổi dữ liệu gốc (raw data), cần sử dụng phương thức `.copy()`:

```python
hotel_df = dfs["hotel_bookings"].copy()
```

Khi đó, `hotel_df` trở thành một bản sao độc lập của dữ liệu gốc. Mọi thao tác làm sạch dữ liệu chỉ ảnh hưởng tới `hotel_df` mà không làm thay đổi `dfs["hotel_bookings"]`.

Việc sử dụng `.copy()` giúp:

* Bảo toàn dữ liệu gốc để đối chiếu khi cần.
* Đảm bảo kết quả EDA luôn được thực hiện trên dữ liệu nguyên bản.
* Tránh các lỗi phát sinh do vô tình thay đổi dữ liệu trong quá trình tiền xử lý.
* Tách biệt rõ ràng giữa dữ liệu thô (raw data) và dữ liệu đã làm sạch (clean data).


````python
hotel_df = dfs["hotel_bookings"].copy()
customers_df = dfs["customers"].copy()
geolocation_df = dfs["geolocation"].copy()
order_items_df = dfs["order_items"].copy()
payments_df = dfs["payments"].copy()
reviews_df = dfs["reviews"].copy()
orders_df = dfs["orders"].copy()
products_df = dfs["products"].copy()
sellers_df = dfs["sellers"].copy()
category_df = dfs["category_translation"].copy()
````

## Handling Missing Values in Hotel Booking Dataset

Trong quá trình khám phá dữ liệu, bảng `hotel_bookings` xuất hiện một số cột chứa giá trị thiếu (missing values). Các phương pháp xử lý được lựa chọn dựa trên ý nghĩa nghiệp vụ của từng thuộc tính:

* **company**: Cột này có tỷ lệ thiếu rất cao (94.31%), do đó được loại bỏ khỏi tập dữ liệu vì không mang lại nhiều giá trị cho phân tích và mô hình học máy.
* **agent**: Giá trị thiếu được thay thế bằng `0`, biểu thị khách hàng không đặt phòng thông qua đại lý du lịch.
* **country**: Giá trị thiếu được thay thế bằng `"Unknown"` nhằm giữ lại toàn bộ bản ghi và phản ánh rõ các trường hợp không xác định quốc gia.
* **children**: Giá trị thiếu được thay thế bằng `0`, giả định rằng khách hàng không đi cùng trẻ em trong các trường hợp dữ liệu bị thiếu.

Sau khi xử lý, tập dữ liệu không còn giá trị thiếu ở các cột trên và sẵn sàng cho các bước phân tích và xây dựng mô hình tiếp theo.


````python

hotel_df.drop(columns=["company"], inplace=True)

hotel_df["agent"] = hotel_df["agent"].fillna(0)

hotel_df["country"] = hotel_df["country"].fillna("Unknown")

hotel_df["children"] = hotel_df["children"].fillna(0)
````

## Handling Missing Values in Review Dataset

Bảng `olist_order_reviews_dataset` chứa hai cột văn bản là `review_comment_title` và `review_comment_message` với tỷ lệ giá trị thiếu khá cao. * `No Message` đối với cột `review_comment_message`

Vì vậy, các giá trị thiếu trong ba cột này được giữ nguyên để đảm bảo tính chính xác và toàn vẹn của dữ liệu.

## Handling Missing Values in Product Dataset

Các giá trị thiếu trong bảng sản phẩm được xử lý dựa trên bản chất của từng thuộc tính.

* `product_category_name` được thay thế bằng `"Unknown Category"` nhằm biểu diễn các sản phẩm chưa xác định được danh mục.
* `product_name_lenght`, `product_description_lenght` và `product_photos_qty` được thay thế bằng giá trị `0`, phản ánh trường hợp sản phẩm không có tên, không có mô tả hoặc không có hình ảnh.
* Các thuộc tính vật lý của sản phẩm gồm `product_weight_g`, `product_length_cm`, `product_height_cm` và `product_width_cm` được thay thế bằng giá trị trung vị (Median). Phương pháp này giúp giảm ảnh hưởng của các giá trị ngoại lai và duy trì phân bố dữ liệu gần với thực tế.

Sau khi xử lý, tập dữ liệu không còn giá trị thiếu và sẵn sàng cho các bước phân tích và xây dựng mô hình học máy.


````python
products_df['product_category_name'] = (
    products_df['product_category_name']
    .fillna('Unknown Category')
)


products_df['product_name_lenght'] = (
    products_df['product_name_lenght']
    .fillna(0)
)

products_df['product_description_lenght'] = (
    products_df['product_description_lenght']
    .fillna(0)
)

products_df['product_photos_qty'] = (
    products_df['product_photos_qty']
    .fillna(0)
)

physical_cols = [
    'product_weight_g',
    'product_length_cm',
    'product_height_cm',
    'product_width_cm'
]

for col in physical_cols:
    products_df[col] = products_df[col].fillna(
        products_df[col].median()
    )
````

## Explanation of Duplicate Count Changes After Data Cleaning

Trong quá trình tiền xử lý dữ liệu, số lượng bản ghi trùng lặp (duplicate records) của bảng `hotel_bookings` thay đổi từ **31,994** lên **32,001** bản ghi. Sự thay đổi này không phải do dữ liệu mới được tạo ra mà là kết quả của quá trình chuẩn hóa dữ liệu.

Ban đầu, dữ liệu được tải từ file gốc vào dictionary `dfs`:

```python
dfs["hotel_bookings"]
```

Tại thời điểm này, số lượng bản ghi trùng lặp được xác định là:

```text
31,994 duplicate records
```

Để bảo toàn dữ liệu gốc, một bản sao độc lập được tạo bằng phương thức `.copy()`:

```python
hotel_df = dfs["hotel_bookings"].copy()
```

Khi đó:

```text
dfs["hotel_bookings"]  -> 31,994 duplicates
hotel_df              -> 31,994 duplicates
```

Sau đó, các bước làm sạch dữ liệu được thực hiện trên `hotel_df`:

```python
hotel_df.drop(columns=["company"], inplace=True)

hotel_df["agent"] = hotel_df["agent"].fillna(0)

hotel_df["country"] = hotel_df["country"].fillna("Unknown")

hotel_df["children"] = hotel_df["children"].fillna(0)
```

Các thao tác này làm thay đổi cách Pandas nhận diện các bản ghi trùng lặp. Trước khi làm sạch, một số bản ghi có thể khác nhau do:

* Giá trị NULL trong các cột `agent`, `country` hoặc `children`.
* Giá trị khác nhau trong cột `company`.

Sau khi thay thế các giá trị thiếu và loại bỏ cột `company`, một số bản ghi trước đây được xem là khác nhau trở thành hoàn toàn giống nhau trên tất cả các thuộc tính còn lại. Vì vậy, Pandas nhận diện thêm **7 bản ghi trùng lặp**, làm tổng số duplicate tăng từ:

```text
31,994 → 32,001
```

Lưu ý rằng dữ liệu gốc trong `dfs["hotel_bookings"]` không bị thay đổi nhờ sử dụng `.copy()`. Do đó:

```python
dfs["hotel_bookings"].duplicated().sum()
```

vẫn trả về:

```text
31,994
```

trong khi:

```python
hotel_df.duplicated().sum()
```

trả về:

```text
32,001
```

Đối với bước xử lý duplicate, giá trị cần sử dụng là **32,001**, vì đây là số lượng bản ghi trùng lặp của tập dữ liệu sau khi đã hoàn thành quá trình làm sạch. Đây cũng là tập dữ liệu sẽ được sử dụng cho các bước phân tích dữ liệu, trực quan hóa và xây dựng mô hình Machine Learning tiếp theo.
